```python
dfm= pd.read_csv('madre_190813.csv', sep=',')
```

```python
dfm.columns
```

```python
dfm['lat_center']= dfm['lat']-np.mean(dfm['lat'])
dfm['lon_center']= dfm['lon']-np.mean(dfm['lon'])
```

```python
def lat_a_m(x):
    """Latitude: 1 deg = 100.54 km"""
    return x*110540
def lon_a_m(x, cos_mean_lat):
    """Longitud: 1 deg = 111.320*cos(latitude) km"""
    return x*111320*cos_mean_lat
```

```python
cos_m_lat=np.cos(np.deg2rad(np.mean(dfm['lat'])))
print(cos_m_lat)
```

```python
dfm['lat_metros'] = dfm['lat_center'].apply(lambda x: round(lat_a_m(x)))
dfm['lon_metros'] = dfm['lon_center'].apply(lambda x: round(lon_a_m(x,cos_m_lat)))
```

```python
dfm.head()
```

```python
dfm.columns
```

```python
dummies_agno=pd.get_dummies(dfm.agno, prefix=None, prefix_sep='_', drop_first=True, sparse=True)

dummies_dia=pd.get_dummies(dfm.dia, prefix=None, prefix_sep='_', drop_first=True,sparse=True)

dummies_mes=pd.get_dummies(dfm.mes, prefix='mes' , prefix_sep='_', drop_first=True, sparse=True)

dummies_cant_universidad=pd.get_dummies(dfm.cant_universidad, prefix='uni', prefix_sep='_', drop_first=True, sparse=True)

dummies_cant_esc=pd.get_dummies(dfm.cant_esc, prefix='esc', prefix_sep='_', drop_first=True, sparse=True)

dummies_cant_tren=pd.get_dummies(dfm.cant_tren, prefix='tres', prefix_sep='_', drop_first=True, sparse=True)

dummies_hora_simple=pd.get_dummies(dfm.hora_simple, prefix='hora', prefix_sep='_', drop_first=True, sparse=True)

dummies_rango_minutos=pd.get_dummies(dfm.rango_minutos, prefix=None, prefix_sep='_', drop_first=True, sparse=True)

dummies_cant_subtes=pd.get_dummies(dfm.cant_subtes, prefix='subte', prefix_sep='_', drop_first=True, sparse=True)
```

```python
dfm=pd.concat([dfm,dummies_agno,dummies_dia,dummies_mes,dummies_rango_minutos,dummies_cant_universidad,dummies_cant_esc,dummies_cant_tren,dummies_hora_simple,dummies_cant_subtes], axis=1)
```
